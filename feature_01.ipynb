## Diseño Evolutivo de Redes Neuronales Artificiales
### feature_01
### Procedimiento para obtener datos de la fuente

El objetivo de esta etapa es describir el procedimiento seleccionado para obtener los datos de las series financieras a usarse en el proyecto.

En la página [Yahoo Finance](https://finance.yahoo.com/) se introduce el ticker de la acción que se desea buscar.

Eso nos lleva por ejemplo para SPY a la [Página de SPY](https://finance.yahoo.com/quote/SPY?p=SPY&.tsrc=fin-srch)

Al darle click a [Historical Data](https://finance.yahoo.com/quote/SPY/history?p=SPY) nos lleva a los datos históricos de SPY donde podemos seleccionar el período deseado y usar el botón de [Download](https://query1.finance.yahoo.com/v7/finance/download/SPY?period1=1559430602&period2=1591053002&interval=1d&events=history) para bajar los datos como csv. El archivo de datos tiene 7 columnas: Date, Open, High, Low, Close, Adj Close y Volume que corresponden a fecha, fecha de apertura, precio máximo, precio mínimo, precio de cierre, precio de cierre ajustado por dividendos y volúmen negociado.

NB: Es importante notar que para poder calcular promedios móviles y otros indicadores técnicos es necesario tener datos anteriores a la fecha donde se desea empezar a entrenar. Por ejemplo para un promedio móvil con ventana de 10 días se necesitan 9 días anteriores a la primera fecha útil, para CCI (Comodity Channel Index) de 20 días se necesitan 39 días anteriores, etc.
