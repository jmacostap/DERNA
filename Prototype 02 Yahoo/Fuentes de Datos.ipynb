### Fuentes de datos
Se crearon varias hojas de cálculo con los datos de Yahoo Finance para usarlas
como datos de prueba.

DJI.csv : Datos de Dow Jones Industrial Average desde 20-jun-1985 hasta el 29-may-2020 bajados a mano directamente de Yahoo Finance.

DJIext.odt : Datos del DJI extendidos con las fórmulas de los indicadores técnicos (ver Indicadores Técnicos.ipynb) con las ventanas más comunes, por ejemplo HH14 es el máximo máximo de 14 días y MA10 es el promedio móvil de 10 días.

GSPC.csv : Datos del Standar & Poor 500 desde el 30-dic-1927 hasta el 1-may-2020 bajados a mano directamente de Yahoo Finance.

GSPCext.odt : Datos del GSPC extendidos igual que DJIext.odt

SPY.csv : Datos del fondo SPY desde el 29-ene-1993 al 01-may-2020
