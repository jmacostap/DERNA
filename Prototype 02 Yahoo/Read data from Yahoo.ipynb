El objetivo de este programa es leer los datos básicos de un identificador de bolsa (Ticker) de la página web de Yahoo Finance y crear un archivo csv con la fecha, precio de apertura, mínimo, máximo, cierre y volumen de cada día disponible.

The objective of this program is to read the basic data of a ticker from the Yahoo Finance website and create a csv file with Date, Open, Min, Max, Close, and Volume for each day for the available date range.

```python
import yfinance as yf

# Este es el único parámetro del programa
# ticker is the only parameter required
ticker="QQQ"
yfticker=yf.Ticker(ticker)
# lee los datos/reads the data from Yahoo finance
historical = yfticker.history(period="max", actions=False)
# crea el archivo csv/creates the csv file
historical.to_csv(ticker+'.csv')
```
