## Diseño Evolutivo de Redes Neuronales Artificiales
### feature01

Procedimiento para obtener datos de la fuente

En la página [Yahoo Finance](https://finance.yahoo.com/) se introduce el ticker de la acción que se desea buscar.

Eso nos lleva por ejemplo para SPY a la [Página de SPY](https://finance.yahoo.com/quote/SPY?p=SPY&.tsrc=fin-srch)

Al darle click a [Historical Data](https://finance.yahoo.com/quote/SPY/history?p=SPY) nos lleva a los datos históricos de SPY donde podemos seleccionar el período deseado y usar el botón de [Download](https://query1.finance.yahoo.com/v7/finance/download/SPY?period1=1559430602&period2=1591053002&interval=1d&events=history) para bajar los datos como csv.
